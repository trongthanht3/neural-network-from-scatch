```python
import pandas as pd
import os
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def pword(text, number):
    texts = text.split('/')
    text = texts[0] + ', ' + 'run_{}'.format(number) + '/' + 'validation'
    return text
```

```python
adam = pd.read_csv('logs/logAdam.csv')
adam.run = adam.run.apply(lambda x: pword(x, 1))
# adam = adam[:21]
adam.head()
```

```python
adagrad = pd.read_csv('logs/logAdagrad.csv')
adagrad.run = adagrad.run.apply(lambda x: pword(x, 2))
# adagrad = adagrad[:21]
adagrad.head()
```

```python
rmsprop = pd.read_csv('logs/logRMSprop.csv')
rmsprop.run = rmsprop.run.apply(lambda x: pword(x, 3))
# rmsprop = rmsprop[:21]
rmsprop.head()
```

```python
sgd = pd.read_csv('logs/logSGD.csv')
sgd.run = sgd.run.apply(lambda x: pword(x, 4))
# sgd = sgd[:21]
sgd.head()
```

```python
df_final = adam.append(adagrad).append(rmsprop).append(sgd)
df_final.reset_index(inplace=True, drop=True)
df_final
```

```python
df_final.info()
```

```python
sns.lineplot(data=df_final, x="epoch", y="epoch_accuracy")
```

```python
# Filter the DataFrame to only validation data, which is what the subsequent
# analyses and visualization will be focused on.
dfw_validation = df_final[df_final.run.str.endswith("/validation")]
dfw_validation
```

```python
# Filter the DataFrame to only validation data, which is what the subsequent
# analyses and visualization will be focused on.
dfw_validation = df_final[df_final.run.str.endswith("/validation")]
# Get the optimizer value for each row of the validation DataFrame.
optimizer_validation = dfw_validation.run.apply(lambda run: run.split(",")[0])

plt.figure(figsize=(20, 8))
plt.subplot(1, 5, 1)
sns.lineplot(data=dfw_validation, x="epoch", y="epoch_accuracy",
             hue=optimizer_validation, markers=True).set_title("accuracy")
plt.subplot(1, 5, 2)
sns.lineplot(data=dfw_validation, x="epoch", y="epoch_loss",
             hue=optimizer_validation).set_title("loss")
plt.subplot(1, 5, 3)
sns.lineplot(data=dfw_validation, x="epoch", y="epoch_precision",
             hue=optimizer_validation).set_title("precision")
plt.subplot(1, 5, 4)
sns.lineplot(data=dfw_validation, x="epoch", y="epoch_recall",
             hue=optimizer_validation).set_title("recall")
plt.subplot(1, 5, 5)
sns.lineplot(data=dfw_validation, x="epoch", y="epoch_f1",
             hue=optimizer_validation).set_title("f1")
```
